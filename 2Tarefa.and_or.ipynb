```python
#1 -Escreva um programa que peça ao usuário inserir três números e, em seguida, determine se um 
#dos três números é maior que a soma dos outros dois.

n1 = int(input("insira o primerio número: "))
n2 = int(input("insira o segundo número : "))
n3 = int(input("insira o tercerio número: "))

if n1 + n2 < n3 or n2 + n3 < n1 or n1 + n3 < n2:
    print(" existe um número que é maior que a soma dos outros dois")
else:
    print("não existe um número que é maior que a soma dos outros dois")
```

```python
#2- Escreva um programa que peça ao usuário inserir duas notas (entre 0 e 100) e determine se o aluno
#passou ou não. Um aluno passa se a soma das notas for maior ou igual a 60 e nenhuma nota é menor que 40.

nota1 = float(input("insira a primerio nota : "))
nota2 = float(input("insira a segundo nota : "))

if nota1 + nota2 >= 60 and nota1> 40 and nota2>40:
    print(" Parabéns, você foi aprovado")
else:
    print(" Você não foi aprovado")
```

```python
#3 -Verifique se uma pessoa é maior ou igual a 18 anos ou se ela tem autorização dos pais.

idade = int(input("Digite qual a sua idade: "))
autoriz = str(input("Você tem autorização dos seus pais? Digite sim ou não: "))

if idade >= 18 or autoriz == "sim":
    print("Você pode entrar na festa")
else:
    print("Você não pode esntrar na festa")
```

```python
#4 Escreva um código que verifique se um número é par ou divisível por 3.

numero = int(input("Entre com valor de número inteiro "))

if numero % 2 == 0:
    print("O número inserido é par")
elif numero % 3 == 0:
    print(" o número inserido é divisível por 3")
else:
    print(" O número inserido não é par e nem divisível por 3") 
```

```python
#6Criar um programa que verifica se uma pessoa tem desconto em um produto, baseado na idade 
#e sua categoria (estudante, aposentado, etc.) e no dia da semana. (Use quantas categorias desejar)

idade = int(input("Digite sua idade: "))
categoria = input("Digite sua categoria (estudante, aposentado, etc.): ")
dia = input("Digite o dia da semana (segunda, terça, etc.): ")

desconto = 0

if idade >= 60:
    desconto = 0.3
elif categoria == "estudante" and (dia == "segunda" or dia == "terça"):
    desconto = 0.2
elif categoria == "aposentado":
    desconto = 0.15
elif categoria == "professor" and dia == "quarta":
    desconto = 0.25

if desconto > 0:
    print("Você tem direito a um desconto de {:.0%}.".format(desconto))
else:
    print("Você não tem direito a nenhum desconto.")
```
