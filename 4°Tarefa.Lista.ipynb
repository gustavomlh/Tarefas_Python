```python
#10 Escreva um programa que peça ao usuário para digitar uma lista de números e, 
#em seguida, imprima a lista com os números duplicados removidos.

lista_duplicados = []
lista_sem_duplicados = []

for i in range(5):
    
    valores = int(input("entre com diferentes valores de numeros"))
    lista_duplicados.append(valores)

while lista_duplicados:
    elemento = lista_duplicados.pop(0) #remove o último elemento de lista_duplicados e retorna esse elemento dentro da variavel elementro 
    if elemento not in lista_duplicados: # se o elemento não estiver dentro de lista duplicadas quer dizer que todas suas cópias foram apagadas
        lista_sem_duplicados.append(elemento) # dessa forma ele fica apto a ser adcionado a lista sem duplicadas 

print("A lista sem duplicados é: " , lista_sem_duplicados)

    
```

```python
#15 Escreva um programa que peça ao usuário para digitar uma lista de números e,
# em seguida, remova todos os números duplicados da lista e imprima a nova lista.

lista_numeros_duplicadas = []
lista_sem_duplicadas = []

for i in range (5):
    valores = int(input("Entre com diferentes numeros: "))
    lista_numeros_duplicadas.append(valores)

while lista_numeros_duplicadas:
    elemento = lista_numeros_duplicadas.pop()
    if elemento not in lista_numeros_duplicadas:
        lista_sem_duplicadas.append(elemento)

lista_sem_duplicadas.sort()
print(lista_sem_duplicadas)

```

```python
#16 Escreva um programa que peça ao usuário para digitar uma lista de números e, em seguida,
# crie uma nova lista que contenha apenas os números que aparecem mais de uma vez na lista original.

lista_numeros_duplicadas = []
lista_proposital = [] # aqui são as duplicadas que colocamos aqui propositalmente 

for i in range (5):
    valores = int(input("Entre com diferentes numeros: "))
    lista_numeros_duplicadas.append(valores)

while lista_numeros_duplicadas:
    elemento = lista_numeros_duplicadas.pop()
    if elemento in lista_numeros_duplicadas: #se o elemento estiver dentro da lista duplicadas nós vamos separar esse que é duplicada de forma proposital na lista proposital porque quer dizer que este elemento é repetido 
        lista_proposital.append(elemento)

lista_proposital.sort()
print(lista_proposital)
```

```python
#18 - Escreva um programa que peça ao usuário para digitar uma lista de números e, 
#em seguida, crie uma nova lista que contenha apenas os números que são divisíveis por 3.


num = [] 

for i in range (5):

    valores = int(input('Digite uma sequencia de números o {}° némero é ' .format(i)))
    if (valores%3) == 0:
        num.append(valores)

num.sort()
print(num)
```

```python
#19- Escreva um programa que peça ao usuário para digitar uma lista de números e, em seguida,
# crie uma nova lista que contenha apenas os números que são divisíveis por um número fornecido pelo usuário.

lista_toda = []
num = []
divisiveis = [] 

divisor = int(input('Digite um número que será o divisor '))
for i in range (5):

    valores = int(input('Digite uma sequencia de números o {}° némero é ' .format(i)))
    lista_toda.append(valores)
    if (valores % divisor) == 0:
        divisiveis.append(valores)
    else:
        num.append(valores)

print("A lista toda é; " , lista_toda, "-> Os numeros divisiveis por" , divisor, "são :" , divisiveis)
```
